```python
# Matplotlib in Python
# Matplotlib is a python library used for data vizulization
# You can create bar-plots, scatter-plots, histograms, and a lot more
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
x = np.arange(1,11)
x
```

```python
y = x*2
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
# adding label and titles
```

```python
plt.plot(x,y)
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.show()
```

```python
plt.plot(x,y,color='g',linestyle=':',linewidth=2)
plt.show()
```

```python
# adding two lines in same plot
```

```python
x = np.arange(1,11)
y1 = x*2
y2 = x*3
```

```python
plt.plot(x,y1,color='g',linestyle=':',linewidth='2')
plt.plot(x,y2,color='r',linestyle='-.',linewidth='3')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.show()
```

```python
# adding sub plot
```

```python
x = np.arange(1,11)
y1 = x*2
y2 = x*3
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,color='blue',linestyle=':',linewidth='2')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.subplot(1,2,2)
plt.plot(x,y2,color='orange',linestyle=':',linewidth='2')
plt.title("Line Plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# bar plot
```

```python
student = {"zain":57,"waseem":24,"akram":83}
```

```python
names = list(student.keys())
values = list(student.values())
```

```python
plt.bar(names,values)
plt.title("Distribution of St. Marks")
plt.xlabel("student names")
plt.ylabel("student marks")
plt.grid(True)
plt.show()
```

```python
# horizental bar plot
```

```python
plt.barh(names,values)
plt.title("Distribution of St. Marks")
plt.xlabel("student marks")
plt.ylabel("student names")
plt.show()
```

```python
# scatter plot
```

```python
x = np.arange(10,70,10)
x
```

```python
from numpy import random

y = random.randint(10, size=(6))
y
```

```python
plt.scatter(x,y)
plt.show()
```

```python
# changing mark asthetics
```

```python
plt.scatter(x,y,marker='*',c='g',s=50)
plt.show()
```

```python
# adding to marker in the same class
```

```python
z = random.randint(10, size=(6))
```

```python
z
```

```python
plt.scatter(x,y,marker='*',c='g',s=50)
plt.scatter(x,z,marker='.',c='r',s=50)
plt.show()
```

```python
# adding subplot

plt.subplot(1,2,1)
plt.scatter(x,y,marker='*',c='g',s=50)

plt.subplot(1,2,2)
plt.scatter(x,z,marker='.',c='r',s=50)
plt.show()
```

```python
# box-plot
```

```python
one = [1,2,3,4,5,5,12]
two = [3,4,14,5,6,5,7]
three = [12,4,5,7,2,3,2]

data = list([one,two,three])
```

```python
plt.boxplot(data)
plt.show()
```

```python
# violin plot
```

```python
plt.violinplot(data,showmedians=True)
plt.show()
```

```python
# pie chart
```

```python
fruit = ["orange","banana","apple","grapes"]
quantity = [23,45,63,15]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
plt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['yellow','blue','green','red'])
plt.show()
```

```python
# doughnut-chart
```

```python
plt.pie(quantity,labels=fruit,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
