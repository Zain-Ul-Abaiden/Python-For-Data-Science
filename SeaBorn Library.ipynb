```python
# SeaBorn plot line
```

```python
import seaborn as sns
```

```python
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri.head()
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
plt.show()
```

```python
# grouping data with hue
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri, hue="event")
plt.show()
```

```python
# adding style
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri, hue="event", style="event")
plt.show()
```

```python
# adding marker
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri, hue="event", style="event", markers=True)
plt.show()
```

```python
# Seaborn histogram / DistPlot
```

```python
diamonds = pd.read_csv("diamonds.csv")
diamonds.head()
```

```python
sns.distplot(diamonds['price'])
plt.show()
```

```python
sns.histplot(diamonds['price'])
plt.show()
```

```python
sns.distplot(diamonds['price'],hist=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color="orange")
plt.show()
```

```python
sns.displot(diamonds['price'],color='green',bins=10,kde=False)
plt.show()
```

```python
sns.distplot(diamonds['price'],color="purple",vertical=True)
plt.show()
```

```python
# seaborn jointplot
```

```python
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
sns.jointplot(x='sepal.length',y='petal.length',data=iris)
plt.show()
```

```python
sns.jointplot(x='sepal.length',y='petal.length',data=iris,color='olive')
plt.show()
```

```python
sns.jointplot(x='sepal.length',y='petal.length',data=iris,color='purple',kind='reg')
plt.show()
```

```python
# seaborn boxplot
```

```python
churn = pd.read_csv("churn.csv")
churn.head()
```

```python
sns.boxplot(x="Churn",y="tenure",data=churn)
plt.show()
```

```python
sns.boxplot(x="InternetService",y="MonthlyCharges",data=churn)
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn, palette="Set1")
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn, palette="Set1", linewidth=3)
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn, order=["Two year", "Month-to-month", "One year"])
plt.show()
```

```python
sns.boxplot(x="Contract",y="tenure",data=churn, hue="PaymentMethod")
plt.show()
```
